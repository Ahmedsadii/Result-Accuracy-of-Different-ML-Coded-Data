# Uni Variate Analysis

```python
df['Type'].value_counts().plot(kind='bar')
```

# Bi Variate Analysis

```python
df[df['Ph.D. granting'] == 'Yes']['Division'].value_counts().plot(kind='bar') #Most division have Ph.D
```

```python
df[df['Ph.D. granting'] == 'No']['Location'].value_counts().head(10).plot(kind='pie',autopct='%1.0f%%' ) #Most Location don't have Ph.D
```

```python
df[df['Type'] == 'Public']['Location'].value_counts().plot(kind='bar')
```

```python
df[df['Type'] == 'Private']['Location'].value_counts().plot(kind='bar')
```

```python
df[df['Type'] == 'Public']['Specialization'].value_counts().plot(kind='bar')
plt.title("Counting all Specialization from Type(Public) data set")
```

```python
df[df['Type'] == 'Private']['Specialization'].value_counts().plot(kind='bar')
plt.title("Counting all Specialization from Type(Private) data set")
```
